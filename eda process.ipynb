# 3. Multivariate analysis

### Combination of 3 categorical columns are represented using lines

In the exploratory data analysis (EDA) process, I have included three multivariate categorical columns: "Location," "Posted Ago," and "Start Date." This strategic choice of variables enables us to identify opportunities where you can quickly find and secure job appointments in a specific city.

Here's how these columns contribute to our analysis:

1. **Location:** The "Location" column categorizes job postings based on their geographical location, such as cities or regions. This allows us to pinpoint cities of interest where you are looking for job opportunities.

2. **Posted Ago:** The "Posted Ago" column categorizes job postings based on when they were posted, indicating their recency. This information helps identify recent job listings, which are often more relevant and likely to result in quicker appointments.

3. **Start Date:** The "Start Date" column provides insights into when the job positions are scheduled to begin. This can be essential in understanding the urgency of job openings and whether they align with your desired timeline for starting a new position.

By combining these three categorical columns, we can perform the following analyses:

- Identify cities with a high frequency of recent job postings, indicating a potentially active job market.
- Filter job postings based on recency, allowing you to focus on opportunities that have been posted recently.
- Match job listings with start dates that align with your preferred timeline, enabling you to quickly find and apply for positions that suit your schedule.

Overall, the inclusion of these multivariate categorical columns enhances our EDA process, empowering us to pinpoint cities with abundant recent job postings. This strategic approach can facilitate a more efficient job search, increasing your chances of securing appointments quickly in a city of your choice.

```python
sns.lineplot(x="Location",y="Posted Ago",hue='Start Date',data=df)
plt.xticks([])
plt.show()
```

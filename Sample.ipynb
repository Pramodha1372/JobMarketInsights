```python
import requests
import re
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://internshala.com/jobs/'
page = requests.get(url)
soup = BeautifulSoup(page.text)
```

```python
page
```

```python
position=[]
for i in soup.find_all('h3',class_='heading_4_5 profile'):
    text=i.text
    text=text.strip()
    if text:
        position.append(text)
    else:
        position.append(np.nan)
```

```python
position
```

```python
len(position)
```

```python
company_name=[]
for i in soup.find_all('h4',class_='heading_6 company_name'):
    text=i.text
    text=text.strip()
    if text:
        company_name.append(text)
    else:
        company_name.append(np.nan)
```

```python
company_name
```

```python
len(company_name)
```

```python
location=[]
for i in soup.find_all('p',id='location_names'):
    text=i.text
    text=text.strip()
    if text:
        location.append(text)
    else:
        location.append(np.nan)
```

```python
location
```

```python
len(location)
```

```python
start_date=[]
for i in soup.find_all('div',id='start-date-first'):
    text=i.text
    text=text.strip()
    if text:
        start_date.append(text)
    else:
        start_date.append(np.nan)
```

```python
start_date
```

```python
len(start_date)
```

```python
ctc=[]
for i in soup.find_all('div',class_='other_detail_item salary_container'):
    text=i.find('div',class_='item_body salary')
    text=text.find('span',class_='desktop')
    text=text.text
    text=text.strip()
    if text:
        ctc.append(text)
    else:
        ctc.append(np.nan)
```

```python
ctc
```

```python
len(ctc)
```

```python
experience=[]
for i in soup.find_all('div',class_='other_detail_item job-experience-item'):
    text=i.find('div',class_='item_body desktop-text')
    text=text.text
    text=text.strip()
    if text:
        experience.append(text)
    else:
        experience.append(np.nan)
```

```python
experience
```

```python
len(experience)
```

```python
posted_ago=[]
for i in soup.find_all('div',class_='posted_by_container'):
    text=i.text
    text=text.strip()
    if text:
        posted_ago.append(text)
    else:
        posted_ago.append(np.nan)
```

```python
posted_ago
```

```python
len(posted_ago)
```

```python
role=[]
for i in soup.find_all('div',class_='other_label_container'):
    text=i.text
    text=text.strip()
    if text:
        role.append(text)
    else:
        role.append(np.nan)
```

```python
role
```

```python
len(role)
```

```python
dic = {
    'Position':position,
    'Company Name':company_name,
    'Location':location,
    'Start Date':start_date,
    'CTC (Annual)':ctc,
    'Experience':experience,
    'Posted Ago':posted_ago,
    'Role':role
}
```

```python
df=pd.DataFrame(dic)
```

```python
df
```
